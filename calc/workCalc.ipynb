```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.backends.backend_pdf import PdfPages
```

```python
workSpaceTrajectoryFile = "../trajectory_generator/logs/WorkSpaceTraj.log"
wSTFile = open(workSpaceTrajectoryFile, "r")
wSTData = []

lines = wSTFile.read().split('\n')
arrayLength = len(lines[0].split('\t'))
miss = 0;
for line in lines:
    data = line.split('\t')
    if len(data) != arrayLength:
        miss = miss + 1
        continue
    wSTData.append(map(float, data))
wSTData = np.array(wSTData).T

X = wSTData[0:3]
t = wSTData[3]
workSpaceFig = plt.figure(2, figsize=(15, 4))
plt.subplot(131)
plt.title("x")
plt.grid(True)
plt.plot(t, X[0])
plt.subplot(132)
plt.title("y")
plt.grid(True)
plt.plot(t, X[1])
plt.subplot(133)
plt.title("z")
plt.grid(True)
plt.plot(t, X[2])
plt.show()
```
